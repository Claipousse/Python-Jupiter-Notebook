```python
#Importation du dataset pour le reste du TP
import numpy as np
data = np.genfromtxt('/kaggle/input/dataset-joueurs/dataset_joueurs.csv',
                     delimiter=',',
                     names = True,
                     dtype = None,
                     encoding = 'utf-8')
```

```python
#Question 2
archers = data[data["Class"] == "Archer"]
meilleur_archer = archers[archers['Nb_Victoires'] == archers['Nb_Victoires'].max()]
print(meilleur_archer["Pseudo"])

#On peut aussi faire comme cela:

#index_archers = np.where(data["Class"] == "Archer")
#archers = data[index_archers]
#index_meilleur_archer = np.argmax(archers["Nb_Victoires"])
#meilleur_archer = archers[index_meilleur_archer]
#print(meilleur_archer["Pseudo"])
```

```python
#Question 3
equipes = np.unique(data["Team"])
print(equipes)
```

```python
#Question 4
equipes = data["Team"]
victoires = data["Nb_Victoires"]

equipes_uniques = np.unique(equipes)
total_victoires_par_equipes = np.array([np.sum(victoires[equipes == equipe])for equipe in equipes_uniques])

index_meilleure_equipe = np.argmax(total_victoires_par_equipes)
meilleure_equipe = equipes_uniques[index_meilleure_equipe]
max_nb_victoires = total_victoires_par_equipes[index_meilleure_equipe]

#Affichage des résultats
print("La meilleure equipe est :", meilleure_equipe)
print("Avec un total de",max_nb_victoires, "victoires.")
```

```python
#Question 5
meilleur_joueur = data[data['Nb_Victoires'] == data['Nb_Victoires'].max()]
print(meilleur_joueur["Pseudo"])
```

```python
#Question 6
classes = np.unique(data["Class"])
print(classes)
```

```python
#Question 7
nb_parties = data["Nb_Parties"]
nb_victoires = data["Nb_Victoires"]
correlation = np.corrcoef(nb_parties,nb_victoires)
print(correlation)
```

```python
#Question 8
classes_uniques = np.unique(data["Class"])
moyenne_victoires = {
    classe: round(data[data["Class"] == classe]["Nb_Victoires"].mean(),2)  #Round permet d'arrondir les résultats au centième près
    for classe in classes_uniques #classe est une variable itératif, à chaque itération, on associe a une classe sa moyenne de victoire 
}

#Affichage des résulats
print("Moyenne de victoires en fonction des classes :")
for classe in classes_uniques:
    print(classe,":",moyenne_victoires[classe])
```

```python
#Question 9
import matplotlib.pyplot as plt

#Données
nb_parties = data["Nb_Parties"]
nb_victoires = data["Nb_Victoires"]

#Création du graphique de dispertion
plt.scatter(nb_parties, nb_victoires)

#Configuration des axes
plt.xlabel("Nombre de parties jouées")
plt.ylabel("Nombre de victoires")
plt.title("Relation entre les parties jouées et le nombre de victoires")

#Affichage
plt.show()
```
